## Resultados

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/tierra.csv')
print(df)
```

```python
media = df.apply(lambda x: x.mean())
print('Media python:', media.Python)
print('Media cython:', media.Cython)
```

```python
print(f'Cython es {media.Python/media.Cython} veces mas rapido que Python')
```

###Graficas

```python
sns.barplot(data=df)
plt.ylabel('Tiempo Segundos')
```

```python
sns.scatterplot(data=df)
plt.ylabel('Tiempo Segundos')
```

#Conclusiones



*   Al cambiar la estructura del lenguaje de python a cython, Cython Demuestra que es casi 25 veces mas eficiente que python en la ejcucion de el mismo programa.

* En la segunda grafica (dispersión) vemos como los tiempos al ser ejecutados en cython mantiene  una tendencia poco variable, lo que por el contrario no pasa en python donde se alcanzan a ver dos saltos que sobrepasan lo 2 segundos de ejecucion, en concreto la muestra 6 y 9. Dandonos a concluir que cython no solo presneta una mejor media de rendimiento si no tambien una mayor estabilidad de tiempos de ejecución. 

* Cython es una herramienta que permite tener resultados sumamente satisfactorios cuando hablamos de optimizacion de tiempos de ejecucion. Reuniendo así lo mejor de dos mundos, la Sintaxis sencilla de python, Con la capacidad e Optimizacion de C.
